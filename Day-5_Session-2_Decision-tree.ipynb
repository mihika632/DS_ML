#### import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('banknotes.csv')
```

```python
df
```

```python
df.shape
```

### Separate the input and output data

```python
# x = df[['Variance','Skewness','Curtosis','Entropy']]
x = df.drop('Class',axis = 1)
y = df['Class']
```

### Explore the data

```python
sns.countplot(x=y)
```

```python
y.value_counts()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.pairplot(df, hue='Class')
```

#### Cross validation

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0, stratify=y)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y.value_counts(0)/len(y)*100
```

```python
y_test.value_counts(0)/len(y_test)*100
```

### Decision Tree Model

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(random_state=0)
```

```python
classifier.fit(x_train, y_train)
```

### Plot the tree

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(16,12))
plot_tree(classifier, fontsize=7,filled=True, class_names=['O','F'],feature_names=x.columns.tolist(), precision=2);
```

# performance evaluation

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics  import classification_report, accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay, confusion_matrixion_matrix
```

```python
ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

```python
335/343
```

```python
print(classification_report(y_test, y_pred))
```
