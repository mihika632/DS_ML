```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[151], [174], [138], [186], [128], [134], [178], [163], [152], [132]])
y = np.array([63,81,56,91, 47, 56, 78, 72, 62, 48])
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x,y)
```

```python
plt.xlabel('Height')
plt.ylabel('weight')
plt.scatter(x,y)
plt.grid()
```

```python
regressor.predict([[185]])
```

```python
regressor.predict([[195]])
```

```python
y_pred = regressor.predict(x)
```

```python
plt.xlabel('Height')
plt.ylabel('weight')
plt.scatter(x,y, color = 'b', label = 'Actual')
plt.plot(x, y_pred, color = 'g', label = 'Predicted')
plt.grid()
plt.legend()
```

```python
regressor.score(x,y)
```
