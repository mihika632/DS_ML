```python
##Separate input and output values

x=df.drop('CLASS',axis=1)
y=df['CLASS']
```

```python
le=LabelEncoder()
```

```python
##Outlier Detection and removal
```

```python
from sklearn.ensemble import IsolationForest
```

```python
iso=IsolationForest(contamination=0.05,random_state=0)
```

```python
clean=iso.fit_predict(x,y)
```

```python
x_clean=x[clean==1]
y_clean=y[clean==1]
```

```python
x_clean.shape,y_clean.shape
```

```python
###  Feature selection or Dimension reduction
```

```python
from sklearn.feature_selection import SelectKBest,chi2
```

```python
skf=SelectKBest(k=5,score_func=chi2)  ##top 5 features will be selected from 20
```

```python
skf.fit(x_clean,y_clean)
```

```python
skf.get_support()
```

```python
x_clean=x_clean.iloc[:,skf.get_support()]
```

```python
x_clean
```

```python
##  Feature scaling

from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_scaled=scaler.fit_transform(x_clean)
```

```python
pd.DataFrame(x_scaled).describe()
```

```python
##cross validation

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_clean,random_state=0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
#3Build the model

from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(random_state=0)
```

```python
clf.fit(x_train,y_train)
```

```python
##  Performance Analysis

from sklearn.metrics import classification_report,accuracy_score
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
y_pred=clf.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
