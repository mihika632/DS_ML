```python
import numpy as n
import matplotlib.pyplot as plt
```

```python
x = n.array([12,23,34,45,56])
y = n.array([15,21,37,42,39])
z = n.array([87,75,54,36,25])
```

# LINEPLOT

```python
plt.title('Line PLot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x,y, color = 'b', marker = 'o', mfc = 'y', linestyle = '-', linewidth = 2, markersize = 1, label = '2022');
plt.plot(x,z, color = 'r', marker = 'o', mfc = 'y', linestyle = '-', linewidth = 2, markersize = 1, label = '2020');
plt.legend();
plt.grid()
#Marker => >, < , o, ^, D,v
```

# subplot

```python
plt.figure(figsize = (16,9))
plt.subplot(2,2,1)
plt.title('Line PLot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.grid()
plt.plot(x,z, color = 'r', marker = 'o', mfc = 'y', linestyle = '-', linewidth = 2, markersize = 1, label = '2020');


plt.subplot(2,2,2)
plt.title('Line PLot')
plt.xlabel('First Value')
plt.ylabel('Second Value')
plt.plot(x,z, color = 'r', marker = 'o', mfc = 'y', linestyle = '-', linewidth = 2, markersize = 1, label = '2020');

plt.grid()
#Marker => >, < , o, ^, D,v
```

# barplot

```python
courses = ['Java', 'Cpp','Python', 'Kotlin']
students = [67, 70, 105, 50]
```

```python
plt.title('STUDENT ENROLLMENT')
plt.ylabel('no of students')
plt.xlabel('courses')
mycolors = ['lightgreen', 'violet', 'cyan', 'yellow', 'green']
plt.grid()
# plt.axis('off')
plt.bar(courses, students, color = mycolors, width = .75, edgecolor = 'black', lw = 1, ls = 'dashed', alpha = 1);
```

# scatterplot

```python
plt.scatter(x, y, marker='o')
plt.title('STUDENT ENROLLMENT')
plt.ylabel('no of students')
plt.xlabel('courses')
```

# pie chart

```python
students
```

```python
plt.pie(students, labels = courses);
```

```python
plt.title('PieChart')
mycolors = ['gold', 'lightgreen', 'cyan', 'yellow']
plt.pie(students, labels = courses, colors = mycolors, autopct = '%.2f%%', shadow = True, explode =(0,0,0,0.2), labeldistance = 0.8, startangle = 90 , normalize=True);
```
